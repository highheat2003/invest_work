```python
import pandas as pd
import numpy as np
import ffn

#  차트 설정
%matplotlib inline
import matplotlib.pyplot as plt

pd.options.mode.chained_assignment = None
```

```python
df = pd.read_csv(r'D:\stock_data\^kospi_d.csv',parse_dates=['Date'])
df.set_index('Date', inplace=True)
df.head()
```

```python
df_month = df.loc[df.groupby(df.index.to_period('M')).apply(lambda x: x.index.max())]
df_month['change_1'] = df_month['Close'].pct_change(1)
df_month['change_2'] = df_month['Close'].pct_change(2)
df_month['change_3'] = df_month['Close'].pct_change(3)
df_month['change_4'] = df_month['Close'].pct_change(4)
df_month['change_5'] = df_month['Close'].pct_change(5)
df_month['change_6'] = df_month['Close'].pct_change(6)
df_month['change_7'] = df_month['Close'].pct_change(7)
df_month['change_8'] = df_month['Close'].pct_change(8)
df_month['change_9'] = df_month['Close'].pct_change(9)
df_month['change_10'] = df_month['Close'].pct_change(10)
df_month['change_11'] = df_month['Close'].pct_change(11)
df_month['change_12'] = df_month['Close'].pct_change(12)
df_month.fillna(0,inplace=True)
df_month
```

```python
df_month3 = df_month.loc[:, 'change_1':'change_12']
for i in range(1,len(df_month3.columns)+1):    
    #print(i)    
    df_month3['val_' + str(i)] = df_month3['change_' + str(i)].apply(lambda x: 1 if x > 0 else 0) 
```

```python
df_month3['val_avg'] = df_month3.loc[:, 'val_1':'val_12'].sum(axis=1)/12
df_month3
```

```python
df_month4 = df_month3[['change_1','val_avg']]
df_month4
```

```python
init_stock = 100
init_cash = 0
cash_rate = 0.001
buy_and_hold = 100

df_month4['return_bh'] = 0
df_month4['return'] = 0
df_month4['stock'] = 0
df_month4['cash'] = 0

# 매월 rebalace
for index,row in df_month4.iterrows():
    # 이전달 금액에서 수일률 적용
    init_stock = init_stock*(row['change_1'] + 1)  
    init_cash = init_cash * (cash_rate + 1)
    buy_and_hold = buy_and_hold*(row['change_1'] + 1)
    # 데이터프레임에 저장
    df_month4.loc[index,'return_bh'] = buy_and_hold
    df_month4.loc[index,'return'] = init_stock + init_cash
    df_month4.loc[index,'stock'] = (init_stock + init_cash)*row['val_avg']
    df_month4.loc[index,'cash'] = (init_stock + init_cash) - df_month4.loc[index,'stock'] 
    # 다음 step으로 값전달    
    init_stock = df_month4.loc[index,'stock']
    init_cash = df_month4.loc[index,'cash']    
    buy_and_hold = df_month4.loc[index,'return_bh']
```

```python
df_month4
```

```python
ax = df_month4[['return','return_bh']].plot(figsize=(16,8),grid=True)
ax.legend(["Time Frame Momentum", "Buy & Holding"])
```
