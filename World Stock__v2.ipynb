```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import yahooFinance as yf
```

```python
def get_stockmarket_data(market):    
    df = yf.getSymbolData(market['code'],sDate=(2017,1,1))
    return df
```

'AxesSubplot' object does not support indexing
https://stackoverflow.com/questions/45993370/matplotlib-indexing-error-on-plotting?rq=1

```python
markets = [{'name':'KOSPI','code':'^KS11'},
           {'name':'Kosdaq','code':'^KOSDAQ'},
           {'name':'S&P 500','code':'^GSPC'},
           {'name':'10year bonds','code':'^TNX'},
           {'name':'Vix','code':'^VIX'},
           {'name':'DAX','code':'^GDAXI'},
           {'name':'Nikkei 225','code':'^N225'},
           {'name':'HANG SENG INDEX','code':'^HSI'},
           {'name':'Jakarta Composite Index','code':'^JKSE'}
          ]

f, axarr = plt.subplots(len(markets), figsize=(16,16), squeeze=False, sharex=False, ncols=2)
f.tight_layout()


row = 0; col = 0;idx = 0
for market in markets:
    market_data = get_stockmarket_data(market)
    #axarr[idx,0].plot(market_data.index, market_data[['Close','SMA50','SMA200']])
        
    if idx%2 == 0:
        col = 0
    else:
        col = 1
                
    axarr[row,col].plot(market_data.index, market_data['close'])
    axarr[row,col].set_title(market['name'])    
    #print(row,col)
    idx = idx + 1
    
    if col == 1:
        col = 0
        row = row + 1        

```
