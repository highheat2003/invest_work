# 데이터 로딩

```python
df = pd.read_csv(r'D:\stock_data\^spx_d.csv').sort_values(by=['Date'],ascending=True)
df.index = pd.to_datetime(df['Date'])
df = df.drop(['Date'], axis=1)
df.head()
```

```python
df.tail()
```

```python
# 월말 종가 데이터 조회
df_month = df.loc[df.groupby(df.index.to_period('M')).apply(lambda x: x.index.max())]
df_month.tail()
```

```python
df_month2 = df_month[df_month.index > '1990-01-01']
df_month2.count()
```

```python
df_month2['Close'].plot(figsize = (16,8))
```

```python
df_month['change_01'] = df['Close']/df['Close'].shift(1) - 1
df_month['change_02'] = df['Close']/df['Close'].shift(2) - 1
df_month['change_03'] = df['Close']/df['Close'].shift(3) - 1
df_month['change_04'] = df['Close']/df['Close'].shift(4) - 1
df_month['change_05'] = df['Close']/df['Close'].shift(5) - 1
df_month['change_06'] = df['Close']/df['Close'].shift(6) - 1
df_month['change_07'] = df['Close']/df['Close'].shift(7) - 1
df_month['change_08'] = df['Close']/df['Close'].shift(8) - 1
df_month['change_09'] = df['Close']/df['Close'].shift(9) - 1
df_month['change_10'] = df['Close']/df['Close'].shift(10) - 1
df_month['change_11'] = df['Close']/df['Close'].shift(11) - 1
df_month['change_12'] = df['Close']/df['Close'].shift(12) - 1
df_month
```

```python
df_month3 = df_month.loc[:, 'change_01':'change_12']
for i in range(0,len(df_month3.columns)):
    tmp = i+1
    column_idx = ""
    if len(str(tmp)) == 1:
        column_idx = "0"+str(tmp)
    else:
        column_idx = str(tmp)
    #print(column_idx)
    
    df_month3['val_' + column_idx] = df_month3['change_' + column_idx].apply(lambda x: 1 if x > 0 else 0)   
```

```python
df_month3['val_avg'] = df_month3.loc[:, 'val_01':'val_12'].sum(axis=1)/12
df_month3
```

```python
df_month4 = df_month3[['change_01','val_avg']]
df_month4
```

```python
pd.options.mode.chained_assignment = None
```

```python
init_stock = 100
init_cash = 0
cash_rate = 0.001
buy_and_hold = 100
#c = {'return': 0,'stock': 0, 'cash': 0}
#df_month4.assign(**c)
df_month4['return_bh'] = 0
df_month4['return'] = 0
df_month4['stock'] = 0
df_month4['cash'] = 0

# 매월 rebalace
for index,row in df_month4.iterrows():
    # 이전달 금액에서 수일률 적용
    init_stock = init_stock*(row['change_01'] + 1)  
    init_cash = init_cash * (cash_rate + 1)
    buy_and_hold = buy_and_hold*(row['change_01'] + 1)
    # 데이터프레임에 저장
    df_month4.loc[index,'return_bh'] = buy_and_hold
    df_month4.loc[index,'return'] = init_stock + init_cash
    df_month4.loc[index,'stock'] = (init_stock + init_cash)*row['val_avg']
    df_month4.loc[index,'cash'] = (init_stock + init_cash) - df_month4.loc[index,'stock'] 
    # 다음 step으로 값전달    
    init_stock = df_month4.loc[index,'stock']
    init_cash = df_month4.loc[index,'cash']    
    buy_and_hold = df_month4.loc[index,'return_bh']
```

```python
ax = df_month4[['return','return_bh']].plot(figsize=(16,8))
ax.legend(["Time Frame Momentum", "Buy & Holding"])
```
