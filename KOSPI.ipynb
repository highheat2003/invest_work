```python
import FinanceDataReader as fdr
import pandas as pd

#  차트 설정
%matplotlib inline
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = 'nanummyeongjo'
plt.rcParams["figure.figsize"] = (14,4)
plt.rcParams['lines.linewidth'] = 2
plt.rcParams["axes.grid"] = True
```

```python
# KOSPI index, 2015~Now
df = fdr.DataReader('KS11', '2015-01-01')
df['SMA50'] = pd.Series(df['Close'].rolling(50).mean(), index=df.index) 
df['SMA200'] = pd.Series(df['Close'].rolling(200).mean(), index=df.index)

df[['Close','SMA50','SMA200']].plot(figsize = (16,8))
```
